```python
#stacking
num_array_1 = np.array([[2,4,5],[5,9,4]])
num_array_2 = np.array([[10,34,21],[45,32,12]])
print("num_array_1 value:",num_array_1)
print("num_array_2 value",num_array_2)
```

```python
print("stacking 2 arrays",np.stack([num_array_1,num_array_2],axis=1))
print("horizontal stacking 2 arrays",np.hstack([num_array_1,num_array_2]))
print("vertical stacking 2 arrays",np.vstack([num_array_1,num_array_2]))
print("depth stacking 2 arrays",np.dstack([num_array_2,num_array_1]))
```
